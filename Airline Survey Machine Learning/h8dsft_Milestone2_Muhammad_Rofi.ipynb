## *8.2 OUTLIERS HANDLING*

- Two techniques are employed for handling outliers, specifically the Z-score method and Tukey's rule.

- The Z-Score method is applicable when the data follows a normal distribution, while Tukey's rule is suitable for skewed data.

- The determination of the appropriate method hinges on the assessment of data skewness.

- If the skewness measure falls within the range of -0.5 to 0.5, the data is indicative of a normal distribution.

- Conversely, if the skewness measure exceeds 0.5 or falls below -0.5, it signifies a skewed data distribution

```python
# SEPERATE BETWEEN DATA DISTRIBUTED NORMAL AND SKEWED
data_normal = []
data_skewed = []
for i in X_train.columns :
    skew = X_train[i].skew()
    if (skew >=0.5) | (skew <=-0.5):
        data_skewed.append(i)

    elif (skew<0.5) & (skew >=-0.5):
        data_normal.append(i)
```

```python
# SPLIT THE DATA WITHIN NORMAL AND SKEWED DISTRIBUTION
print(f'DATA WITH NORMAL DISTRIBUTION :{data_normal}')
print(f'DATA WITH SKEWED DISTRIBUTION :{data_skewed}')
```

```python
# FUNCTION FOR OUTLIER HANDLING

# DATA WITH NORMAL DISTRIBUTION USE Z-SCORE METHOD (GAUSSIAN)
m_gaussian = Winsorizer(capping_method='gaussian',
                             tail='both',
                             fold=3,
                             variables= data_normal)

m_gaussian = m_gaussian.fit_transform(X_train[data_normal])

# DATA WITH NORMAL DISTRIBUTION USE TUKEYS RULE METHOD (IQR)
m_iqr = Winsorizer(capping_method='iqr',
                             tail='both',
                             fold=1.5,
                             variables= data_skewed)

m_iqr = m_iqr.fit_transform(X_train[data_skewed])
```

```python
# REJOIN COLUMNS THAT HAVE MANAGED OUTLIERS
X_train = pd.concat([m_gaussian, m_iqr],axis = 1)
X_train.head(3)
```

# CONCEPTUAL PROBLEMS

1. Latar Belakang dan Cara Kerja Bagging:
   - Latar belakang Bagging: Bagging (Bootstrap Aggregating) dikembangkan untuk mengatasi overfitting dalam model machine learning. Overfitting terjadi ketika model terlalu kompleks dan tidak generalisasi dengan baik.
   - Cara kerja Bagging: Bagging bekerja dengan membuat sejumlah sampel data pelatihan acak (bootstrap) dan melatih banyak model dengan sampel-sampel ini. Setiap model memberikan prediksi, dan hasil akhir diambil dengan mengagregasi prediksi tersebut, misalnya dengan menggunakan voting atau rata-rata.

2. Perbedaan Cara Kerja Random Forest dan Boosting:
   - Random Forest: Random Forest adalah ensamble learning yang menggunakan beberapa pohon keputusan. Setiap pohon dilatih secara independen, dan prediksi diambil melalui mayoritas suara (voting) dari pohon-pohon tersebut.
   - Boosting: Boosting adalah teknik yang memperkuat model lemah menjadi model yang lebih kuat. Model-model dihasilkan secara berurutan, dengan model baru fokus pada kesalahan model sebelumnya. Prediksi akhir dibuat dengan menggabungkan kontribusi dari semua model.

3. Cross Validation:
   - Cross Validation adalah metode evaluasi kinerja model machine learning yang digunakan untuk mengukur sejauh mana model mampu menggeneralisasi dari data pelatihan ke data yang tidak terlihat. 
   - Cara kerja Cross Validation: Data pelatihan dibagi menjadi sejumlah subset (biasanya k-fold), lalu model dilatih dan diuji pada subset-subset tersebut secara bergantian. Hasilnya digunakan untuk menghitung metrik evaluasi seperti akurasi, presisi, atau recall, sehingga dapat memberikan gambaran yang lebih baik tentang kinerja model.
