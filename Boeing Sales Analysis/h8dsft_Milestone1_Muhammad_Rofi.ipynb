### **DROP KOLOM**

Kolom unfilled orders harus di drop karena memiliki banyak missing value dan tidak memiliki dampak yang signifikan terhadap pengolahan data

```python
# MENGGUNAKAN METHODE 'DROP' ('AXIS=1' MENYATAKAN KOLOM) 
data = dataset.drop(['Unfilled Orders'], axis = 1)
data.info()
```

Kolom Unfilled order berhasil di drop

### **DROP MISSING VALUE**

'''

- Setelah dianalisa lebih lanjut ternyata banyak pesanan pesawat yang gagal kirim. Pesanan gagal kirim ini tidak bisa dimasukan kedalam pengolahan data karena nantinya pengolahan data menjadi tidak valid. Oleh karena itu saya memutuskan untuk mendrop semua pesanan yang mengalami gagal kirim

```python
datafull = data.dropna()

datafull.info()
```

Ternyata pesanan yang berhasil dikirimkan ke pelanggan hanya berjumlah 8024

### **MENGUBAH WHITESPACE PADA NAMA KOLOM MENJADI '_'**

```python
# MENGGUNAKAN METODE LOWER (MENGUBAH DARI UPPERCASE MENJADI LOWER CASE) & REPLACE (UNTUK MENGUBAH WHITESPACE MENJADI '_')
datafull.columns = [col.lower().replace(' ', '_') for col in data.columns]
datafull.columns
```

- Whitespace berhasil dihilangkan dan semua nama column menjadi lower case

- Kolom delivery year masih memiliki nama yang kurang sesuai, maka kita harus ubah kembali

```python
datafull.rename(columns={"delivery_year_" : "delivery_year"} , inplace=True)
datafull.columns
```

```python
# PETA HEATMAP BERDASARKAN JUMLAH PEMBELIAN PESAWAT

map_country_data = dataok[['country','delivery_total']].groupby('country')['delivery_total'].sum().rename('counts')
map_country = px.choropleth(map_country_data, locationmode = 'country names' , locations=map_country_data.index ,color='counts' , hover_data=['counts'] , color_continuous_scale='Jet' , title = 'TOTAL PEMBELIAN PESAWAT BERDASARKAN NEGARA')
map_country.show()
```

```python
dataok.head()
```

```python
# BAGAIMANA DISTRIBUSI JUMLAH PEMBELIAN DI SETIAP NEGARA ?
jumlah_negara = dataok.groupby('country').agg({'delivery_total': 'sum'}).sort_values(by='delivery_total', ascending=False)
print('\n** 20 NEGARA DENGAN JUMLAH PEMBELIAN TERTINGGI **\n')
print(jumlah_negara.head(20)) #AMBIL TOP 20 DENGAN VALUE TERTINGGI

# VISUALISASI
# AMBIL TOP 20 NEGARA DENGAN JUMLAH PEMBELIAN TERTINGGI
top_20_countries = jumlah_negara.head(20)
```

```python
# BERAPA STANDAR DEVIASI DARI JUMLAH PEMBELIAN PESAWAT DI SETIAP NEGARA UNTUK MENGUKUR VARIASINYA ?

std_negara = dataok.groupby('country').agg({'delivery_total': 'std'}).sort_values(by='delivery_total', ascending=False).round(0)
print('\n** 20 NEGARA DENGAN STANDAR DEVIASI JUMLAH PEMBELIAN TERTINGGI **\n')
print(std_negara.head(20)) # AMBIL 20 NEGARA DENGAN NILAI STANDAR DEVIASI TERTINGGI
```

```python
# AMBIL STANDAR DEVIASI UNTUK TOP 20 NEGARA DENGAN SUBSCRIBERS TERTINGGI BERDASARKAN INDEX "TOP_20_COUNTRIES"
std_dev_top_20 = std_negara.loc[top_20_countries.index]

# Visualiasi
plt.figure(figsize=(15, 10))
sns.barplot(x=top_20_countries['delivery_total'], y=top_20_countries.index)
plt.title('20 NEGARA DENGAN JUMLAH PEMBELIAN TERTINGGI')
plt.xlabel('JUMLAH PEMBELIAN')
plt.ylabel('NEGARA')
plt.show()

plt.figure(figsize=(15, 10))
sns.barplot(x=std_dev_top_20['delivery_total'], y=std_dev_top_20.index)
plt.title('STANDAR DEVIASI JUMLAH PEMBELIAN UNTUK TOP 20 NEGARA')
plt.xlabel('STANDAR DEVIASI PEMBELIAN PESAWAT')
plt.ylabel('NEGARA')
plt.show()
```

```python
# NEGARA DENGAN JUMLAH PEMBELIAN TERTINGGI
pembeli_terbanyak = jumlah_negara.idxmax().values[0]
print(f'NEGARA DENGAN PEMBELIAN TERBANYAK : {pembeli_terbanyak}')

# Negara dengan jumlah subscribers terendah
pembeli_sedikit = jumlah_negara.idxmin().values[0]
print(f'NEGARA DENGAN PEMBELIAN TERENDAH : {pembeli_sedikit}')
```

## *PENGARUH MODEL PESAWAT TERHADAP TINGKAT PEMBELIAN PESAWAT*

**PERTANYAAN :** Apakah ada hubungan antara model pesawat dengan tingkat pembelian pesawat ?

**TUJUAN :** Menyediakan informasi kepada stakeholder terkait tentang bagaimana model pesawat dapat mempengaruhi kesuksesan penjualan pesawat serta memprediksi tren global.

Untuk dapat dilakukannya uji hipotesis maka saya akan mengelompokan pesawat berdasarkan tipe body pesawatnya yaitu :

- NARROW BODY AIRCRAFT : pesawat berlorong tunggal yang digunakan pada rute domestik regional

- WIDEBODY AIRCRAFT : pesawat berlorong ganda yang digunakan untuk rute jarak jauh lintas benua

Lalu dibuat tabel baru yang menunjukan index tipe body pesawat. Terakhir baru dilakukan penghitungan korelasi antara tipe body pesawat dengan jumlah pesawat yang dikirim ke pelanggan menggunakan spearman method

```python
# MEMBUAT LIST PATTERN STRING
widebody = '|'.join(["727", "737-200", "737-800", "737-700C", "737-600", "737-700", "MD-80", "737 MAX", "DC-9", "737-300", "737-500", "737-400", "737-700W", "737-800A", "717-200", "MD-90", "737-900ER", "737-900", "737-100"])

# ASSIGN KATEGORI MODEL PESAWAT DENGAN ANGKA 1 / 2 SESUAI DENGAN KATEGORI YANG SUDAH DITENTUKAN.
dataok['tipe_body'] = np.where(dataok['model_series'].str.contains(widebody, case=False) , 1, 2)

dataok.head(15)
```

**KORELASI ANTARA TIPE BODY PESAWAT DENGAN TOTAL PENJUALAN PESAWAT**

```python
corr_rho, pval_s = stats.spearmanr(dataok['delivery_total'], dataok['tipe_body'])
print(f"rho-correlation: {corr_rho:.2f}")
print(f"p-value: {pval_s}")

# VISUALISASI DATA
sns.lmplot(data=dataok,x='delivery_total', y='tipe_body')
plt.show()
```

H0 = Tidak ada hubungan antara Tipe body pesawat dengan Jumlah pembelian pesawat 

H1 = Ada hubungan antara Tipe body pesawat dengan Jumlah pembelian pesawat


- **rho-correlation**:

value: -0.20, menunjukkan korelasi positif yang sangat kuat antara 'tipe body pesawat' dan 'jumlah penjualan'. jika satu variabel meningkat, yang lain memiliki kecenderungan untuk meningkat juga.

- p-value (5.066447557842344e-72):

P-value ini adalah angka yang sangat kecil, jauh < dari 0,05, maka dianggap signifikan secara statistik. Dengan kata lain, sangat tidak mungkin hubungan yang kita lihat hanyalah sebuah kebetulan. 

**KESIMPULAN :**
Terdapat hubungan yang KUAT antara jumlah pembelian pesawat dengan tipe body pesawat (pesawat berlorong tunggal atau pesawat berbadan lebar). Semakin kecil kapasitas penumpang pesawatnya maka ada kemungkinan pesawatnya akan semakin diminati. Hal ini dapat diartikan sebagai tren minat pesawat global saat ini adalah pesawat bertipe NARROW BODY
